## Imports and setup

```python
import numpy as np
from SimPEG import Utils
from zephyr import *
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%pylab inline
```

```python
import matplotlib
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png')
matplotlib.rcParams['savefig.dpi'] = 150 # Change this to adjust figure size

# Plotting options
font = {
    'family': 'Bitstream Vera Sans',
    'weight': 'normal',
    'size': 12,
}

matplotlib.rc('font', **font)
```

## Configuration

```python
cellSize    = 1             # m
freq        = 1e2           # Hz        (should this be multiplied by 2*pi?)
velocity    = 2000          # m/s
density     = 1             # units of density
Q           = 2e1           # can be inf
nx          = 100           # count
nz          = 200           # count
freeSurf    = [False, False, False, False] # t r b l
dims        = (nx,nz)       # tuple
sLoc        = [ 10,190]     # x, z
c           = np.ones(dims) * velocity
rho         = np.ones(dims) * density
systemConfig = {
    'dx':   cellSize,       # m
    'dz':   cellSize,       # m
    'freq':     freq,       # Hz
    'c':        c,          # m/s
    'rho':      rho,        # density
    'Q':        Q,          # can be inf
    'nx':       nx,         # count
    'nz':       nz,         # count
    'sLoc':     sLoc,       # x, z
    'freeSurf': freeSurf,   # t r b l
}
```

## Mesh + system initialization

```python
mesh, A, Ainv = initHelmholtzNinePointCE(systemConfig)
```

## Source initialization

```python
q       = np.zeros(mesh.nN)
qI      = Utils.closestPoints(mesh, sLoc, gridLoc='N')
q[qI]   = 1
```

## Testing

```python
%%time
u       = Ainv * q
```

```python
amax    = abs(u).max()

ax1 = plt.subplot(1,2,1, aspect=1)
plt.set_cmap(cm.gray_r)
im1 = mesh.plotImage(q, vType='N', ax=ax1)
plt.title('Source location')

ax2 = plt.subplot(1,2,2, aspect=1)
plt.set_cmap(cm.bwr)
im2 = mesh.plotImage(u, vType='N', ax=ax2)
im2[0].set_clim(-amax, amax)
plt.title('Wavefield')
```
