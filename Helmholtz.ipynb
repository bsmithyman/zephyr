## Imports and setup

```python
import numpy as np
import SimPEG
from SimPEG import Utils
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%pylab inline
```

```python
import matplotlib
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png')
matplotlib.rcParams['savefig.dpi'] = 150 # Change this to adjust figure size

# Plotting options
font = {
    'family': 'Bitstream Vera Sans',
    'weight': 'normal',
    'size': 12,
}

matplotlib.rc('font', **font)
```

## Configuration

```python
cellSize    = 1         # m
freq        = 5e1       # Hz        (should this be multiplied by 2*pi?)
velocity    = 2000      # m/s
Q           = 2e1       # can be inf
nx          = 100       # count
nz          = 200       # count
sLoc        = [ 10,190] # x, z
```

## Mesh + system initialization

```python
hx      = np.ones(nx) * cellSize
hz      = np.ones(nz) * cellSize
mesh    = SimPEG.Mesh.TensorMesh([hx, hz], '00')
omega   = 2 * np.pi * freq

if Q is inf:
    c   = np.ones(mesh.nC) * velocity               #   (nx * nz)   by 1
else:
    c   = np.ones(mesh.nC) * (velocity - (1j * velocity) / (2*Q)) 

L       = mesh.nodalLaplacian                       # (nx+1 * nz+1) by (nx+1 * nz+1)
mAve    = mesh.aveN2CC                              #   (nx * nz)   by (nx+1 * nz+1)
k2      = Utils.sdiag(mAve.T * (omega**2 / c**2))
A       = L + k2
```

```python
fig = plt.figure()
kwargs = {
    'cmap':             cm.gray,
    'interpolation':    'none',
    #'vmin':             -4,
    #'vmax':             1,
}

ax1 = fig.add_subplot(1,3,1, aspect=1)
plt.imshow(k2[:5,:5].todense().real, **kwargs)
plt.title('Diagonal term')

ax2 = fig.add_subplot(1,3,2, aspect=1)
plt.imshow(L[:5,:5].todense(), **kwargs)
plt.title('Laplacian')

ax3 = fig.add_subplot(1,3,3, aspect=1)
plt.imshow(A[:5,:5].todense().real, **kwargs)
plt.title('System matrix')
```

## Source initialization

```python
q       = np.zeros(mesh.nN)
qI      = Utils.closestPoints(mesh, sLoc, gridLoc='N')
q[qI]   = 1
```

## Testing

```python
%%time
Ainv    = SimPEG.SolverLU(A)
u       = Ainv * q
```

```python
amax    = abs(u).max()

ax1 = plt.subplot(1,2,1, aspect=1)
plt.set_cmap(cm.gray_r)
im1 = mesh.plotImage(q, vType='N', ax=ax1)
plt.title('Source location')

ax2 = plt.subplot(1,2,2, aspect=1)
plt.set_cmap(cm.bwr)
im2 = mesh.plotImage(u, vType='N', ax=ax2)
im2[0].set_clim(-amax, amax)
plt.title('Wavefield')
```
