## System / modelling configuration

```python
cellSize    = 1             # m
freqs       = [2e2]         # Hz
velocity    = 2500          # m/s
density     = 2700          # units of density
Q           = 50           # can be inf
nx          = 256           # count
nz          = 256           # count
freeSurf    = [False, False, False, False] # t r b l
dims        = (nx,nz)       # tuple
nPML        = 32
c           = np.fliplr(np.ones(dims) * velocity + np.linspace(0,4000,nz).reshape((1,nz)))
rho         = np.fliplr(np.ones(dims) * density)
nfreq       = len(freqs)    # number of frequencies
nky         = 48            # number of y-directional plane-wave components
nsp         = nfreq * nky   # total number of 2D subproblems

srcs        = np.array([[128,0,128], [64,0,192]])
recs        = np.array(zip(np.linspace(0, 255, 128), np.zeros((128,)), np.ones((128,))*128))
recmode     = 'fixed'

geom        = {
    'src':  srcs,
    'rec':  recs,
    'mode': 'fixed',
}

# Base configuration for all subproblems
systemConfig = {
    'dx':   cellSize,       # m
    'dz':   cellSize,       # m
    'c':        c.T,        # m/s
    'rho':      rho.T,      # density
    'Q':        Q,          # can be inf
    'nx':       nx,         # count
    'nz':       nz,         # count
    'freeSurf': freeSurf,   # t r b l
    'nPML':     nPML,
    'geom':     geom,
}

subConfigSettings = {
    'freqs':    freqs,
    'nky':      nky,
    'cmin':     c.min(),
}
# Generate the configuration update objects, which contain freq, ky and weight
forward25D = getForward(systemConfig, subConfigSettings)
```

## Examples

```python
%%time
# Forward model 2.5D wavefield for a given source and reduce to local array
isrc = 0
wavefield0, data0 = forward25D(isrc)
```

```python
%%time
# Forward model 2.5D wavefield for a given source and reduce to local array
isrc = 1
wavefield1, data1 = forward25D(isrc)
```

```python
clip = max(abs(wavefield0)) * 1e-1

fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
plt.imshow(wavefield0.reshape((nz+1,nx+1)).real, cmap=cm.bwr, vmin=-clip, vmax=clip)
ax1.invert_yaxis()
ax2 = fig.add_subplot(1,2,2)
plt.imshow(wavefield1.reshape((nz+1,nx+1)).real, cmap=cm.bwr, vmin=-clip, vmax=clip)
ax2.invert_yaxis()
```

```python
fig = plt.figure()

scal0 = max(abs(data0))
ax1 = fig.add_subplot(1,2,1, projection='3d')
ax1.scatter(np.arange(128), data0.real, data0.imag, s=1)
ax1.set_xlim([0, 128])
ax1.set_ylim([-2*scal0, 2*scal0])
ax1.set_zlim([-2*scal0, 2*scal0])

scal1 = max(abs(data1))
ax2 = fig.add_subplot(1,2,2, projection='3d')
ax2.scatter(np.arange(128), data1.real, data1.imag, s=1)
ax2.set_xlim([0, 128])
ax2.set_ylim([-2*scal1, 2*scal1])
ax2.set_zlim([-2*scal1, 2*scal1])
```
