# Configuration

```python
cellSize    = 1             # m
freqs       = [2e2]         # Hz
velocity    = 2500          # m/s
density     = 2700          # units of density
Q           = inf           # can be inf
nx          = 256           # count
nz          = 256           # count
freeSurf    = [True, False, False, False] # t r b l
dims        = (nx,nz)       # tuple
nPML        = 32
c           = fliplr(np.ones(dims) * velocity + np.linspace(0,4000,nz).reshape((1,nz)))
rho         = fliplr(np.ones(dims) * density)
nfreq       = len(freqs)    # number of frequencies
nky         = 40            # number of y-directional plane-wave components
nsp         = nfreq * nky   # total number of 2D subproblems

# Base configuration for all subproblems
systemConfig = {
    'dx':   cellSize,       # m
    'dz':   cellSize,       # m
    'c':        c.T,        # m/s
    'rho':      rho.T,      # density
    'Q':        Q,          # can be inf
    'nx':       nx,         # count
    'nz':       nz,         # count
    'freeSurf': freeSurf,   # t r b l
    'nPML':     nPML,
}

# Additional configuration terms that vary between subproblems
genSubConfigs = lambda freqs: [{'freq': freq, 'ky': ky} for freq in freqs for ky in np.linspace(0, freq / c.min(), nky)]

# Weights according to whether a subproblem is the zero wavenumber or not
genWeights = lambda subConfigs: [1 if item['ky'] == 0 else 2 for item in subConfigs]
```

```python
# Set up dictionary for subproblem objects
dview['localSystem'] = {}

# Push base configuration for the system
dview['baseSystemConfig'] = systemConfig

# Create a function to get a subproblem object by tag and reference it locally
dview['systemFromTag'] = lambda tag: localSystem[tag]
systemFromTag = Reference('systemFromTag')

# Create a function to get a subproblem forward modelling function
dview['forwardFromTag'] = lambda tag, loc: systemFromTag(tag).forward(loc)
forwardFromTag = Reference('forwardFromTag')

# Generate the configuration update objects, which contain freq, ky and weight
subConfigs = genSubConfigs(freqs)
weights = genWeights(subConfigs)
for i in xrange(nsp):
    subConfigs[i].update({'kyweight': weights[i]})

# Set up the subproblem objects with each new configuration
tags = dview.map_sync(setupSystem, subConfigs)

# Forward model in 2.5D (in parallel) for an arbitrary source location
forward25D = lambda loc: reduce(add, dview.map(forwardFromTag, tags, [loc]*nsp))
```

```python
%%time
# Forward model 2.5D wavefield for a given source and reduce to local array
sLoc = [128, 128]
wavefield0 = forward25D(sLoc)
```

```python
%%time
# Forward model 2.5D wavefield for a given source and reduce to local array
sLoc = [64,64]
wavefield1 = forward25D(sLoc)
```
